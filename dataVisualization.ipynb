```python
sns.relplot(x="cylinders", y="mpg", kind="line", col="origin", data=df)
plt.xticks(rotation=90)

plt.show()
```

```python
sns.relplot(x="weight", y="mpg", kind="scatter", col="origin", data=df)
plt.xticks(rotation=90)

plt.show()
```

```python
sns.relplot(x="mpg", y="model year", kind="line", col="origin", data=df)
plt.xticks(rotation=90)

plt.show()
```
